```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

<font color="ORANGERED"> **Load dataset** from sklear.dataset.

```python
housing = fetch_california_housing()
```

<font color="ORANGERED"> First array denotes **input** and second array denotes **target** data.

<font color="ORANGERED"> Explore the dataset clearer with **Pandas DataFrame** object.

```python
pd.DataFrame(data=np.c_[housing.data, housing.target], columns=housing.feature_names+['TARGET'])
```

<font color="ORANGERED"> Split the whole dataset into **train**, **validation**, and **test** data.

```python
x_train, x_test, y_train, y_test = train_test_split(housing.data, housing.target)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train)
print("x_train", x_train.shape, "y_train", y_train.shape)
print("x_val", x_val.shape, "y_val", y_val.shape)
print("x_test", x_test.shape, "y_test", y_test.shape)
```

<font color="ORANGERED"> **Standarize** and **normalize** the whole dataset.

```python
x_train = StandardScaler().fit_transform(x_train)
x_val = StandardScaler().fit_transform(x_val)
x_test = StandardScaler().fit_transform(x_test)
```

```python
x_train = MinMaxScaler().fit_transform(x_train)
x_val = MinMaxScaler().fit_transform(x_val)
x_test = MinMaxScaler().fit_transform(x_test)
```

<font color="ORANGERED"> Perform an advance split into **multiple overlapping input (A and B)**, where **input A** consists of feature 0 to 4 and **input B** consists of feature 2 to 7.

```python
x_train_A, x_train_B = x_train[:, :5], x_train[:, 2:]
x_val_A, x_val_B = x_val[:, :5], x_val[:, 2:]
x_test_A, x_test_B = x_test[:, :5], x_test[:, 2:]
x_new_A, x_new_B = x_test_A[:3], x_test_B[:3] # samples to predict after training
print("x_train_A", x_train_A.shape, "x_train_B", x_train_B.shape)
print("x_val_A", x_val_A.shape, "x_val_B", x_val_B.shape)
print("x_test_A", x_test_A.shape, "x_test_B", x_test_B.shape)
```

## <font color="ORANGERED">**Keras Functional API** - Complex neural network layers are connected *in custom*.</font>

> <font color="red"> **(Step 1) Build the Model**</font>
* Define parameters: weight and bias
* Define hyperparameters: total neurons on (1) input layers, (2) hidden layers, and (3) output layer, also (4) activation functions
* In Keras Functional API, define which layer connects to which layer

<br> See **Keras Core Layers** documentation [**here**](https://keras.io/api/layers/core_layers/).
<br> See **Keras Model.summary() and Model.get_layers() method** documentation [**here**](https://keras.io/api/models/model/#summary-method).
<br> See **Keras Layer Initializer** for weights (*kernel_initializer*) and biases (*bias_initializer*) [**here**](https://keras.io/api/layers/initializers/).
<br> See **Keras Activation Functions** [**here**](https://keras.io/api/layers/activations/).
<br> By default on .layers.Dense(), activation=None, kernel_initializer="glorot_uniform" and bias_initializer="zeros".

```python
input_A = keras.layers.Input(shape=[5])
input_B = keras.layers.Input(shape=[6])
hidden1 = keras.layers.Dense(30, activation='relu')(input_B)
hidden2 = keras.layers.Dense(30, activation='relu')(hidden1)
concat = keras.layers.concatenate([input_A, hidden2])
output = keras.layers.Dense(1, name="output")(concat)
aux_output = keras.layers.Dense(1, name="aux_output")(hidden2)
regression_model = keras.Model(inputs=[input_A, input_B], outputs=[output, aux_output])
regression_model.summary()
```

**Notes:** In Keras,
* bias_initializer="random_normal" is **equivalent** to bias_initializer=tf.keras.initializers.random_normal()
* kernel_initializer="glorot_uniform" is **equivalent** to kernel_initializer=tf.keras.initializers.glorot_uniform()
* activation="relu" is **equivalent** to activation=tf.keras.activations.relu()

Therefore, parameter="x" is **equivalent** to parameter=tf.keras.____.x()
<br> It's also applied for loss function, optimizer, and metrics later.
<br> To adjust or change parameters flexibly, use its original function instead.

> <font color="red"> **(Step 2) Compile the Model** </font>
* Define hyperparameters: loss function, optimizer (with learning rate and momentum), and metrics (optional)

See **Keras Model.compile() method** documentation [**here**](https://keras.io/api/models/model_training_apis/#compile-method).
<br> See **Keras Loss Function** documentation [**here**](https://keras.io/api/losses/).
<br> See **Keras Optimizers** documentation [**here**](https://keras.io/api/optimizers/).
<br> See **Keras Metrics** documentation [**here**](https://keras.io/api/metrics/).

```python
regression_model.compile(
    loss=["mean_squared_error", "mean_squared_error"], # Define two losses due to two generated outputs
    loss_weights=[.9, .1], 
    optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3), # by default Adam optimizer learning_rate=0.001
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

**Notes:** In Keras (for Regression problem),
* loss="mean_squared_error" is **equivalent** to loss=tf.keras.losses.MeanSquaredError()
* optimizer="Adam" is **equivalent** to optimizer=tf.keras.optimizers.Adam()

Therefore, parameter="x" is slightly **equivalent** to parameter=tf.keras.____.x()
<br> To adjust or change parameters flexibly, use its original function instead.
<br> For instance, by default "Adam" will set learning_rate=0.01. To adjust, implement *tf.keras.optimizers.Adam(learning_rate=...)* instead.

> <font color="red"> **(Step 3) Train/Fit the Model** </font>
* Define hyperparameters: epochs and batch_size
* Initialize validation_data (optional)

See **Keras Model.fit() method** documentation [**here**](https://keras.io/api/models/model_training_apis/#fit-method).
<br> For Keras ver. 2.3.0-tf, Model.fit() batch_size=None by default. Therefore, try to set it by 1 to get actual train data.

```python
reg_result = regression_model.fit(
    x=[x_train_A, x_train_B],
    y=[y_train, y_train],
    epochs=20, batch_size=1,
    validation_data=([x_val_A, x_val_B], [y_val, y_val])
)
```

```python
pd.DataFrame(reg_result.history).plot(figsize=(15, 8))
plt.grid(True)
plt.xlim(0, 20)
plt.xlabel("Epochs")
plt.ylabel("Scores")
```

> <font color="red"> **(Step 4) Evaluate the Model with test data** </font>

See **Keras Model.evaluate() method** documentation [**here**](https://keras.io/api/models/model_training_apis/#evaluate-method).

```python
regression_model.evaluate([x_test_A, x_test_B], [y_test, y_test])
```

```python
regression_model.predict([x_new_A, x_new_B])
```

<font color="ORANGERED"> **Callbacks** or **EarlyStopping**

```python
checkpoint = keras.callbacks.ModelCheckpoint("model/regression_model.h5", save_best_only=True)
early_stopping = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
```

```python
reg_result2 = regression_model.fit(
    x=[x_train_A, x_train_B],
    y=[y_train, y_train],
    epochs=20, batch_size=1,
    validation_data=([x_val_A, x_val_B], [y_val, y_val]),
    callbacks=[checkpoint, early_stopping]
)
```

```python
pd.DataFrame(reg_result2.history).plot(figsize=(15, 8))
plt.grid(True)
plt.xlim(0, 20)
plt.xlabel("Epochs")
plt.ylabel("Scores")
```
