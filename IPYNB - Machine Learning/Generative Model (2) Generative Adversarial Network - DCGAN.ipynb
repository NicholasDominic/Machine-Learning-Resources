> **BRIEF INTRODUCTION**<br>
Generative model is capable of randomly generating new data that looks *very similar* to the training data. Simply put, the output is the input itself (same size). Therefore in this case, the output also called as **reconstruction**. Two types of generative model are: **Autoencoders (AE)** and **Generative Adversarial Networks (GAN)**.

## CIFAR-10 Dataset Loading, Normalization, and Split

```python
# Load the dataset from Keras
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
# Convert to floats
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# Scaling from [0, 255] to [-1, 1]
x_train = (x_train - 127.5) / 127.5
x_test = (x_test - 127.5) / 127.5

# Get validation data from train data
x_train, x_val = x_train[:-5000], x_train[-5000:]
y_train, y_val = y_train[:-5000], y_train[-5000:]

# Show the shape of each
print("Image Dimension: {}".format(x_train.shape[1:]))
print("x_train: {} y_train: {}".format(x_train.shape, y_train.shape))
print("x_val: {} y_val: {}".format(x_val.shape, y_val.shape))
print("x_test: {} y_test: {}".format(x_test.shape, y_test.shape))
```

## User-defined Functions

```python
# select real samples
def generate_real_samples(dataset, n_samples):
    ix = randint(0, dataset.shape[0], n_samples) # choose random instances
    x = dataset[ix] # retrieve selected images
    y = ones((n_samples, 1)) # generate "real" class labels (1)
    return x, y
```

```python
# generate points in latent space as input for the generator
def generate_latent_points(latent_dim, n_samples):
    x_input = randn(latent_dim * n_samples) # generate points in the latent space
    x_input = x_input.reshape(n_samples, latent_dim) # reshape into a batch of inputs for the network
    return x_input
```

```python
# use the generator to generate n fake examples, with class labels
def generate_fake_samples(g_model, latent_dim, n_samples):
    x_input = generate_latent_points(latent_dim, n_samples) # generate points in latent space
    x = g_model.predict(x_input) # predict outputs
    y = zeros((n_samples, 1)) # generate "fake" class labels (0)
    return x, y
```

```python
%matplotlib inline
def plot_fake_images(examples, epoch, n=7):
    examples = (examples + 1) / 2.0 # scale from [-1,1] to [0,1]
    for i in range(n * n): # has 7 rows x 7 columns
        plt.subplot(n, n, 1+i)
        plt.axis("off")
        plt.imshow(examples[i])
    plt.show()
```

```python
def summarize_performance(epoch, g_model, d_model, dataset, latent_dim, n_samples=150):
    x_real, y_real = generate_real_samples(dataset, n_samples) # prepare real samples
    _, acc_real = d_model.evaluate(x_real, y_real, verbose=0) # evaluate discriminator on real examples
    
    x_fake, y_fake = generate_fake_samples(g_model, latent_dim, n_samples) # prepare fake examples
    _, acc_fake = d_model.evaluate(x_fake, y_fake, verbose=0) # evaluate discriminator on fake examples
    
    plot_fake_images(x_fake, epoch)
    print("=========================================")
    print("[SUMMARIZE until Epoch {}]\nAccuracy real: {}%, fake: {}%".format(epoch+1, round(acc_real*100, 2), round(acc_fake*100, 2)))
    print("=========================================\n")
```

```python
def fit_model(gan_model, dataset, latent_dim=100, epochs=200, n_batch=128):
    g_model, d_model = gan_model.layers # unfold generator and discriminator layer
    batch_per_epoch = int(dataset.shape[0] / n_batch)
    half_batch = int(n_batch / 2)
    
    # enumerate epochs
    for i in range(epochs):
        # enumerate batches over the training set
        print("Epoch {}/{}".format(i+1, epochs))
        for j in range(batch_per_epoch):
            # get randomly selected real samples
            x_real, y_real = generate_real_samples(dataset, half_batch)
            
            # update discriminator model weights
            d_loss1, _ = d_model.train_on_batch(x_real, y_real)
            
            # generate fake examples
            x_fake, y_fake = generate_fake_samples(g_model, latent_dim, half_batch)
            
            # update discriminator model weights
            d_loss2, _ = d_model.train_on_batch(x_fake, y_fake)
            
            # prepare points in latent space as input for the generator
            x_gan = generate_latent_points(latent_dim, n_batch)
            
            # create inverted labels for the fake samples
            y_gan = ones((n_batch, 1))
            
            # update the generator via the discriminator's error
            g_loss, _ = gan_model.train_on_batch(x_gan, y_gan)
            
            # summarize loss on this batch
            if (j+1) % 39 == 0: # print out in every 39 batches only
                print("\t[Batch {}/{}] Discriminator Loss (Real/Fake): {}/{} - Generator Loss: {}".
                      format(j+1, batch_per_epoch, round(d_loss1, 2), round(d_loss2, 2), round(g_loss, 2)))
        
        # evaluate the model performance every 10 epochs
        if (i+1) % 10 == 0:
            summarize_performance(i, g_model, d_model, dataset, latent_dim)
    
    # save the model
    g_model.save("model/dcgan_model/generator_dcgan.h5")
    d_model.save("model/dcgan_model/discriminator_dcgan.h5")
    gan_model.save("model/dcgan_model/gan.h5")
```

* **Step 1: BUILD**

```python
generator = keras.models.Sequential([
    keras.layers.Dense(256 * 4 * 4, input_dim=100), # number_of_nodes, latent_space
    keras.layers.LeakyReLU(alpha=.2),
    keras.layers.Reshape((4, 4, 256)),

    # upsample to 8x8
    keras.layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),

    # upsample to 16x16
    keras.layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),
    
    # upsample to 32x32
    keras.layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),

    # output layer
    keras.layers.Conv2D(3, (3, 3), activation="tanh", padding="same")
], name="Generator")
generator.summary()
```

```python
discriminator = keras.models.Sequential([
    keras.layers.Conv2D(64, (3,3), padding="same", input_shape=(32, 32, 3)),
    keras.layers.LeakyReLU(alpha=.2),
    
    # downsample
    keras.layers.Conv2D(128, (3,3), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),

    # downsample
    keras.layers.Conv2D(128, (3,3), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),

    # downsample
    keras.layers.Conv2D(256, (3,3), strides=(2, 2), padding="same"),
    keras.layers.LeakyReLU(alpha=.2),
    
    # classifier
    keras.layers.Flatten(),
    keras.layers.Dropout(.4),
    keras.layers.Dense(1, activation="sigmoid")
], name="Discriminator")
discriminator.summary()
```

```python
dcgan = keras.models.Sequential([generator, discriminator], name="DeepConv_GANs")
dcgan.summary()
```

* **Step 2: COMPILE**

```python
discriminator.compile(
    loss="binary_crossentropy",
    optimizer=keras.optimizers.Adam(lr=2e-4, beta_1=.5),
    metrics=['accuracy']
)
discriminator.trainable = False
```

```python
dcgan.compile(
    loss="binary_crossentropy",
    optimizer=keras.optimizers.Adam(lr=2e-4, beta_1=.5),
    metrics=['accuracy']
)
```

* **Step 4: EVALUATE**

```python
fit_model(dcgan, x_train, epochs=100)
```
