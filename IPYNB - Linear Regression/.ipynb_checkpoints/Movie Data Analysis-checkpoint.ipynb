### [1.1] Data Cleaning: Missing/Null Values

```python
dataset.isnull().any() # Check if there's any column which contains null values, marked/returned as TRUE
```

```python
col_has_null = dataset.columns[dataset.isnull().any()] # retrieve column name
dataset[col_has_null].isnull().sum() # retrieve number of rows which has null values

# Another way to show all columns which has NULL value(s)
# for i in col_has_null.tolist():
#     print("Total missing/null value(s) in {}: {}".format(i, dataset[i].isnull().sum()))
```

```python
# Display rows which column has null value(s)
display_null_col = dataset[dataset.isnull().any(axis=1)]
display_null_col
```
