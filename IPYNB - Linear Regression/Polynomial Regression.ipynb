# POLYNOMIAL REGRESSION

```python
from numpy.random import rand, randn
from numpy import linspace
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec as grid
from sklearn.preprocessing import PolynomialFeatures as pf
from sklearn.linear_model import LinearRegression as lr
```

Generate a 100 random numbers of **x**.

```python
x = 6 * rand(100, 1) - 3
x
```

Create the polynomial regression function, **y**.

```python
# y = 0.7x^n + x + 2 + Gaussian error term, where n = 1-10
y = []
for i in range(1, 10+1):
    if i == 1:
        polynome = (.7 * x)
    else:
        polynome = (.7 * x**i) + x + 3 + randn(100, 1)
    y.append(polynome)
```

Graphical form of polynomial regression function.

```python
poly_names = ['Linear', 'Quadratic', 'Cubic', 'Quartic/Biquadratic', 'Quintic', 'Sextic/Hextic', 'Septic/Hectic', 'Octic', 'Nonic', 'Decic']
poly_colors = ['xkcd:bubblegum pink', 'xkcd:golden', 'xkcd:aqua', 'xkcd:deep sky blue', 'xkcd:coral', 'xkcd:lime', 'xkcd:orange', 'xkcd:heliotrope', 'xkcd:spearmint', 'xkcd:desert']
fig = plt.figure(figsize=(20, 5))
for i in range(10):
    a = fig.add_subplot(2, 5, i+1) # .add_subplot(row, column, position) where position must starts from 1
    a.scatter(x, y[i], color=poly_colors[i])
    if i == 0: plt.title(poly_names[i] + "\ny = 0.7x^" + str(i+1) + " err")
    else: plt.title(poly_names[i] + "\ny = 0.7x^" + str(i+1) + " + x + err")
fig.tight_layout() # margins adjusted
```

```python
polynomial = []
for i in range(1, 10+1):
    polynomial.append(pf(degree=i, include_bias=False))
polynomial
```

```python
poly_regression = []
for i in range(1, 10+1):
    poly_regression.append(lr().fit(polynomial[i-1].fit_transform(x), y[i-1])) # i-1 because i starts from 1, not 0
    print("Degree = ", i)
    print("\tIntercept: ", poly_regression[i-1].intercept_) # i-1 because i starts from 1, not 0
    print("\tCoefficients: ", poly_regression[i-1].coef_) # i-1 because i starts from 1, not 0
```

```python
fig = plt.figure(figsize=(15, 20))
x_axis = linspace(-3, 3, 100).reshape(-1, 1) # Generate a set of number starts from -3 to 3

for i in range(10):
    a = fig.add_subplot(5, 2, i+1); # .add_subplot(row, column, position) where position must starts from 1
    a.scatter(x, y[i], color=poly_colors[i])
    a.plot(x_axis, poly_regression[i].predict(polynomial[i].transform(x_axis)), "k--", linewidth=2.5)
    if i == 0: plt.title(poly_names[i] + "\ny = 0.7x^" + str(i+1) + " err")
    else: plt.title(poly_names[i] + "\ny = 0.7x^" + str(i+1) + " + x + err")
fig.tight_layout() # margins adjusted
```
